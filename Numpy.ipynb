#### Flattening a numpy array

Sometimes when you have a multidimensional array and want to collapse it to a single-dimensional array, you can either use the flatten() method or the ravel() method:

```python
a = np.ones((2,2))

b = a.flatten() # Does deep copy, original value wont change.
c = a.ravel() # Does shallow copy, original value will change.

print('Original shape :', a.shape)
print('Array :','\n', a)
print('Shape after flatten :',b.shape)
print('Array :','\n', b)
print('Shape after ravel :',c.shape)
print('Array :','\n', c)
```

But an important difference between flatten() and ravel() is that the former returns a copy of the original array while the latter returns a reference to the original array. This means any changes made to the array returned from ravel() will also be reflected in the original array while this will not be the case with flatten().

```python
print(a, '---', b)
```

```python
# Flatten -> Creates a deep copy which does not change the original array
b[0] = 0
print(a)
print(b)
```

```python
# Ravel -> Creates a Shallow copy which changes the original array
c[0] = 0
print(a)
print(c)
```

## Stacking and Concatenating Numpy arrays

#### Stacking ndarrays

1. vstack() - Combine the arrays vertically (i.e. along the rows) using the vstack() method, thereby increasing the number of rows in the resulting array.

2. hstack() - combine the arrays in a horizontal fashion (i.e. along the columns) using the hstack(), thereby increasing the number of columns in the resultant array

```python
a = np.arange(0,5)
b = np.arange(5,10)
print('Array 1 :','\n',a)
print('Array 2 :','\n',b)
print('Vertical stacking :','\n',np.vstack((a,b)))
print('Horizontal stacking :','\n',np.hstack((a,b)))
```

A point to note here is that the axis along which you are combining the array should have the same size otherwise you are bound to get an error!

#### dstack() method -> depth stack

Another interesting way to combine arrays is using the dstack() method. It combines array elements index by index and stacks them along the depth axis:

```python
a = [[1,2],[3,4]]
b = [[5,6],[7,8]]

c = np.dstack((a,b))

print('Array 1 :','\n',a)
print('Array 2 :','\n',b)
print('Dstack :','\n',c)
print('----')
print(c.shape)
```

#### Concatenating ndarrays

Another way of joining two arrays using concatenate() method in numpy along their particular axis.

```python
a = np.arange(0,5).reshape(1,5)
b = np.arange(5,10).reshape(1,5)
print('Array 1 :','\n',a)
print('Array 2 :','\n',b)
print('Concatenate along rows :','\n',np.concatenate((a,b), axis=0))
print('Concatenate along columns :','\n',np.concatenate((a,b),axis=1))
```

## Min-Max values and their indexes

Min and Max values in an ndarray can be easily found using the min() and max() methods

```python
a = np.array([[1,2, 8],[5,6, 3]])
print(a)
# minimum along a column
print('Min :',np.min(a,axis=0))
# maximum along a row
print('Max :',np.max(a,axis=1))
```

### Argmin and Argmax

Determine the index of the minimum or maximum value in the ndarray along a particular axis using the argmin() and argmax() methods

```python
a = np.array([[1,2,8], [5,6,3]])
print(a)
# minimum along a column
print('Min :',np.argmin(a,axis=0))
# maximum along a row
print('Max :',np.argmax(a,axis=1))
```

## Sorting in Numpy arrays

Numpy uses sort() method to sort the arrays. The sort parameter has an option to sort in many ways like mergesort, heapsort, quicksort and so on.

```python
a = np.array([1,4,2,5,3,6,8,7,9])
np.sort(a)
```

The sorting can also be in desired axis.

```python
a = np.array([[5,6,7,4],
              [9,2,3,7]])# sort along the column
print('Before sort', a)

print('Sort along column :','\n',np.sort(a, kind='mergresort',axis=1))
# sort along the row
print('Sort along row :','\n',np.sort(a, kind='mergresort',axis=0))
```
