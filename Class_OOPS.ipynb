# OOPS

What is OOPS?

Object-oriented programming (OOP) is a computer programming model that organizes software design around data, or objects, rather than functions and logic.


Structure of Object-Oriented Programming Language?

The structure, or building blocks, of object-oriented programming include the following:

* **Classes** are user-defined data types that act as the blueprint for individual objects, attributes and methods.
* **Objects** are instances of a class created with specifically defined data. Objects can correspond to real-world objects or an abstract entity. When class is defined initially, the description is the only object that is defined.

* **Methods** are functions that are defined inside a class that describe the behaviors of an object. Each method contained in class definitions starts with a reference to an instance object. Additionally, the subroutines contained in an object are called instance methods. Programmers use methods for reusability or keeping functionality encapsulated inside one object at a time.

* **Attributes** are defined in the class template and represent the state of an object. Objects will have data stored in the attributes field. Class attributes belong to the class itself.

Fundamental concepts supported by OOPS are:

1. Classes
2. Object
3. Inheritance
4. Polymorphism
5. Encapsulation and many...

```python
# Inheritance

class arithmetic:
    def __init__(self,a,b):
        self.a = a
        self.b = b

    def summation(self):
        print(self.a)
        print(self.b)
        return self.a+self.b

    def division(self,b):
        return b / self.a

    def combine(self):
        a = self.summation()
        b = self.division(45)
        return a*b
```

```python
class student(arithmetic):
    def power(self):
        x = self.summation() # self.a+self.b -> Comes from parent class
        print(x)
        y = self.a * self.b
        print(self.a, self.b)
        print(x)
        return y + x
```

```python
# With inheritance

class Vehicle:

    def __init__(self, name, color, price):
        self.name = name
        self.color = color
        self.price = price

    def show(self):
        print('Details:', self.name, self.color, self.price)

    def max_speed(self):
        print('Vehicle max speed is 150')

    def change_gear(self):
        print('Vehicle change 6 gear')

# inherit from vehicle class
class Car(Vehicle):
    def max_speed(self):
        print('Car max speed is 240')

    def change_gear(self):
        print('Car change 7 gear')


# Car Object
car = Car('Car x1', 'Red', 20000)
car.show()
# calls methods from Car class
car.max_speed()
car.change_gear()

# Vehicle Object
vehicle = Vehicle('Truck x1', 'white', 75000)
vehicle.show()
# calls method from a Vehicle class
vehicle.max_speed()
vehicle.change_gear()
```

## Encapsulation

Encapsulation is the one of the fundamental principles of OOPS. It is a way to restrict the some components of the object.

1. **Public:** Class members without any underscore prefix are considered public and can be accessed from anywhere.
2. **Protected:** Class members with a single underscore prefix, like _variable or _method(), are considered protected. It indicates that the member should not be accessed from outside the class or its subclasses directly, although it is still accessible.
3. **Private:** Class members with a double underscore prefix, like __variable or __method(), are considered private. It indicates that the member should not be accessed or overridden outside the class. Python performs name mangling on private members to make them less accessible

```python
class Hello:
  def __init__(self):
    self.a = 10
    self._b = 20
    self.__c = 30

hello = Hello()

print(hello.a) # Public
print(hello._b) # Protected
print(hello.__c) # Private
```

```python
class Hello:
  def __init__(self, name):
    self.a = 10
    self._b = 20
    self.__c = 'pyhton'

  def public_method(self):
    print(self.a)
    print(self.__c[0:3]) # Here self.__c can be called inside class and not directly
    print('public')

hello = Hello('name')

#print(hello.a)
print(hello._b)
#print(hello.__c)
hello.public_method()
```

```python
### To create method only as private instead of variables

class Hello:
  def __init__(self):
    self.a = 10
    self._b = 20
    self.__c = 30

  def __private_method(self):
    add = 5 + 10
    return add

  def public_method(self):
    print(self.a)
    print(self.__private_method() + self.__c)

hello = Hello()

hello.public_method()
#hello.__private_method()
```

#### Regex functions

1. **re.findall()** -> method returns a list of strings containing all matches.

2. **re.split()** -> method splits the string where there is a match and returns a list of strings where the splits have occurred.

3. **re.sub()** -> syntax: re.sub(pattern, replace, string) : The method returns a string where matched occurrences are replaced with the content of replace variable.

4. **re.search()** -> match = re.search(pattern, str), returns a match object; if not, it returns None.

5. re.match() -> function to search pattern within the test_string.

#### Meta Characters

**[]  ->  Square** brackets specifies a set of characters you wish to match. eg [abc], [a-e], [0-9]

**.  -> Period** matches any single character except newline('\n'). expression -> ..

**^  -> caret** is to check if a symbol starts with character or not.

**$  -> dollar** is to check if a symbol ends with character or not.

(*)(No bracket)  -> star matches zero or more occurrences of the pattern left to it.

**(+)(No bracket)  -> plus** matches one or more occurrences of the pattern left to it.

**?  -> question mark** matches zero or one occurrence of the pattern left to it.

**{}  **-> Consider this code: {n,m}. This means at least n, and at most m repetitions of the pattern left to it.

**| -> Vertical bar** | is used for alternation (or operator).

**() -> Parentheses** () is used to group sub-patterns. For example, (a|b|c)xz match any string that matches either a or b or c followed by xz.

**\ -> Backlash** \ is used to escape various characters including all metacharacters. If you are unsure if a character has special meaning or not, you can put \ in front of it. This makes sure the character is not treated in a special way.

#### Special sequence

1. \d - Matches any decimal digit. Equivalent to [0-9].

2. \D - Matches any non-decimal digit. Equivalent to [^0-9].

3. \w - Matches any alphanumeric character (digits and alphabets). Equivalent to [a-zA-Z0-9_]. By the way, underscore _ is also considered an alphanumeric character.

4. \W - Matches any non-alphanumeric character. Equivalent to [^a-zA-Z0-9_].

### Findall using RegEx

1. Retreive Email_id and phone number

```python
import re
```

```python
chat1 = 'you ask lot of questions 1235678912,abc@xyz.com'
chat2 = 'here it is:(123)-567-8912,abc_80@xyz12.com'
chat3 = 'yes, phone:1235678912 email: abc@xyz.com'
```

```python
# Email id:
pattern = '[A-Za-z0-9_]*@[a-z0-9]*.com'

"""
[a-zA-Z0-9_] -> All chracters, numbers and underscore
* -> All after the extracted characters
"""

match = re.findall(pattern, chat1)
print(match)
print(match[0])
```

```python
# Phone number
pattern = '(\d{10})|(\(\d{3}\)-\d{3}-\d{4})'

"""
\d{10} -> 10 digits
| -> 'Or' symbol
\ -> escape
\(\d{3}\) -> extracts only 3 digits within ()
"""

match_phno = re.findall(pattern, chat2)
print(match_phno)
print(match_phno[0][1])
```

```python
def get_pattern_match(pattern, text):
  matches = re.findall(pattern, text)
  if matches:
      return matches[0]
```

```python
def extract_personal_information(text):
  age = get_pattern_match('age (\d+)', text)
  full_name = get_pattern_match('Born(.*)\n', text)
  birth_date = get_pattern_match('Born.*\n(.*)\(age', text)
  birth_place = get_pattern_match('\(age.*\n(.*)', text)
  return {
      'age': int(age),
      'name': full_name.strip(),
      'birth_date': birth_date.strip(),
      'birth_place': birth_place.strip()
  }
```
