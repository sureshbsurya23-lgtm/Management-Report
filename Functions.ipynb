# Function call inside another function

```python
def fun1(a,b):
  return a*b

def fun2(a,b):
  return a/b

def main_fun(a,b):
  c = fun1(a,b)       # c = 10*20
  d = fun2(a,b)       # d = 10/20
  res = c+d           # res = (10*20) + (10/20)
  return res
```

```python
result = main_fun(10, 20)
result
```

In the above code cell,

squared_numbers = `[(lambda x: x ** 2) for num in numbers]`: This line uses a list comprehension to create a new list called squared_numbers. For each element num in the numbers list, it creates a lambda function lambda x: x ** 2 and adds it to the squared_numbers list. So, the result is nothing but lambda functions in the list comprehension. To get the output, we need to insert the value of `x` for each lambda function, which is nothing but `num` here.

```python
# Example 7: Lambda function in a list comprehension
numbers = [1, 2, 3, 4, 5]
squared_numbers = [(lambda x: x ** 2)(num) for num in numbers]
squared_numbers
```

#Recursive function

Recursion is the process of defining something in terms of itself.


Recursion.JPG

```python
def factorial(x):
    """This is a recursive function
    to find the factorial of an integer"""
    if x == 1:
        return 1
    else:
        return (x * factorial(x-1))

num = 5
print("The factorial of", num, "is", factorial(num))
```

```python
# 1. You should have a end condition at any cost. -> Similar to while loop
```

```python
# # 1. factoriol(5)
#       -> (5 * factorial(4))
#       -> (5 * 4 * factorial(3))
#       -> (5 * 4 * 3 * factorial(2))
#       -> (5 * 4 * 3 * 2 * factorial(1))
#       -> (5 * 4 * 3 * 2 * 1)
```

```python
def recur_sum(n):

  if n <= 1:
      return n
  else:
      return n + recur_sum(n-1)

recur_sum(5)
```

```python
def fibonacci(n):
  if n <= 1:
      return n
  else:
      return fibonacci(n - 1) + fibonacci(n - 2)

n = 15
fibonacci(n)

# for i in range(1,n):
#   print(fibonacci(i))
```
